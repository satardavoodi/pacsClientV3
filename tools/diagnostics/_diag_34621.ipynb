```python
import sqlite3
from pathlib import Path

DB = Path(r'C:\AI-Pacs codes\aipacs-pydicom2d\user_data\database\dicom.db')
conn = sqlite3.connect(str(DB))
conn.row_factory = sqlite3.Row
cur = conn.cursor()

# Find patient 34621
cur.execute('SELECT * FROM patients WHERE patient_id LIKE ?', ('%34621%',))
patient = cur.fetchone()
if patient:
    print(f'FOUND: pk={patient["patient_pk"]}, id={patient["patient_id"]!r}, name={patient["patient_name"]}')
else:
    print('NOT FOUND by patient_id')
    cur.execute('SELECT patient_pk, patient_id, patient_name FROM patients ORDER BY patient_pk')
    for r in cur.fetchall():
        print(f'  pk={r["patient_pk"]}, id={r["patient_id"]!r}, name={r["patient_name"]}')
```
